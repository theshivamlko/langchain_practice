```python
import os
from itertools import chain

from dotenv import load_dotenv
from langchain_core import messages

load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

messages=[
    SystemMessage(content="You are an English to French Expert"),
    HumanMessage(content="Hello World!")
]

result=llm.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser=StrOutputParser()
output_parser.invoke(result)

```

```python
chain = llm | output_parser
chain.invoke(messages)
```

```python
from  langchain_core.prompts import  ChatPromptTemplate

template ="Translate following  into {language}"
prompt = ChatPromptTemplate.from_messages([("system", template),("user","{text}")])

result=prompt.invoke({"language":"French","text":"Hello World!"})
```

```python
result.to_messages()
```

```python
chain = prompt | llm | output_parser
chain.invoke({"language":"French","text":"Hello World!"})
```
